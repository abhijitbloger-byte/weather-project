# WEATHER FORCAST ANALYSIS
-------------------


## Loading DataSet

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Weather Data.csv")
```

## 1. Find all unique Wind Speed values

```python
sp = df["Wind Speed_km/h"].unique()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x="Wind Speed_km/h")

plt.title("Frequency of Wind Speeds")
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Count")
```

## 2. Find the number of times when the weather is exactly "Clear"

```python
pf = df[df["Weather"] == "Clear"]
print(pf)
```

```python
plt.figure(figsize=(12,6))

sns.countplot(
    data=pf,
    x="Weather"
)

plt.title("Number of Clear Weather Records")

plt.show()
```

## 3. Find all instances where Wind Speed is exactly 4 km/h

```python
print(df[df["Wind Speed_km/h"] == 4])
```

```python
plt.figure(figsize=(12,6))

sns.histplot(df["Wind Speed_km/h"], bins=20)

plt.axvline(4, color='red', linestyle='--')

plt.title("Wind Speed Distribution")

plt.show()
```

## 4. Find the mean Visibility

```python
nf = df["Visibility_km"].mean()
print(nf)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Visibility_km"], bins=20, kde=True)

# plt.axvline(
#     nf,
#     color='red',
#     linestyle='--'
# )

plt.title("Visibility Distribution")

plt.show()
```

## 5. Find the standard deviation of Pressure (kPa)

```python
nf = df["Press_kPa"].std()
print(nf)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Press_kPa"], bins=20, kde=True)

# plt.axvline(df["Press_kPa"].mean(),
#             color='green',
#             linestyle='--')

plt.title("Pressure Distribution")

plt.show()
```

## 6. Find all instances when Snow was recorded

```python
sn = df[df["Weather"] == "Snow"]
print(sn)
```

```python
snow = df[df["Weather"].str.contains("Snow")]

plt.figure(figsize=(10,5))

sns.countplot(data=snow,
              y="Weather")

plt.title("Snow Related Weather")

plt.show()
```

## 7. Show all records when Weather Condition is Fog

```python
fog = df[df["Weather"] == "Fog"]
print(fog)
```

```python
# fog = df[df["Weather"]=="Fog"]

plt.figure(figsize=(8,5))

sns.histplot(fog["Visibility_km"],
             bins=15,
             kde=True)

plt.title("Visibility During Fog")

plt.show()
```

## 8. Find the mean value of each column against each Weather Condition

```python
weather_mean = df.groupby("Weather").mean(numeric_only=True)
print(weather_mean)
```

```python
# weather_mean = df.groupby("Weather").mean(numeric_only=True)

plt.figure(figsize=(12,8))

sns.heatmap(weather_mean,
            annot=True,
            cmap="coolwarm",
            fmt=".1f")

plt.title("Mean Weather Statistics")

plt.show()
```

## 9. Find all instances where Weather is Clear OR Visibility is above 40

```python
fa = df[(df["Weather"] == "Clear") | (df["Visibility_km"] > 40)]
print(fa)
```

```python


plt.figure(figsize=(10,6))

sns.plot(data=fa,
                x="Visibility_km",
                y="Temp_C",
                hue="Weather")

plt.title("Clear Weather OR Visibility > 40")

plt.show()
```

## 10. Find all instances where Weather is Clear AND Relative Humidity is above 50

```python
sf = df[(df["Weather"] == "Clear") & (df["Rel Hum_%"] > 50)]
print(sf)
```

```python


plt.figure(figsize=(10,6))

sns.scatterplot(data=sf,
                x="Rel Hum_%",
                y="Temp_C")

plt.title("Clear Weather AND Humidity > 50")

plt.show()
```
